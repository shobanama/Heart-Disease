# Data Cleaning


```
import numpy as np
import pandas as pd
```

```
new_df = pd.read_csv('raw.csv')
```

```
# na count before clean
new_df.isnull().sum()
```

```
def data_cleaning():

# Read new dataset
    df = new_df

#age          		
    df['age'].fillna(df['age'].mean(), inplace=True)

#sex          
    df['sex'].fillna(df['sex'].max(), inplace=True)

#CP 
    df['cp'].fillna(df['cp'].mean(), inplace=True)

#trestbps     
    trestbps_ngtv = df['trestbps'] < 0
    df.loc[trestbps_ngtv, 'trestbps'] = np.nan
    df['trestbps'].fillna(df['trestbps'].mean(), inplace=True)

#chol
    chol_ngtv = df['chol'] < 0
    df.loc[chol_ngtv, 'chol'] = np.nan
    df['chol'].fillna(df['chol'].mean(), inplace=True)

#fbs
    fbs_ngtv = df['fbs'] < 0
    df.loc[fbs_ngtv, 'fbs'] = np.nan
    df['fbs'].fillna(df['fbs'].max(), inplace=True)

#restecg
    restecg_ngtv = df['restecg'] < 0
    df.loc[restecg_ngtv, 'restecg'] = np.nan
    df['restecg'].fillna(df['restecg'].max(), inplace=True)

#Thalach
    thalach_ngtv = df['thalach'] < 0
    df.loc[thalach_ngtv, 'thalach'] = np.nan
    df['thalach'].fillna(df['thalach'].mean(), inplace=True)

#Exang
    exang_ngtv = df['exang'] < 0
    df.loc[exang_ngtv, 'exang'] = np.nan
    df['exang'].fillna(df['exang'].mean(), inplace=True)

#OldPeak
    oldpeak_ngtv = df['oldpeak'] < 0
    df.loc[oldpeak_ngtv, 'oldpeak'] = np.nan
    df['oldpeak'].fillna(df['oldpeak'].mean(), inplace=True)

#Slope
    slope_ngtv = df['slope'] < 0
    df.loc[slope_ngtv, 'slope'] = np.nan
    df['slope'].fillna(df['slope'].mean(), inplace=True)

#Ca
    ca_ngtv = df['ca'] < 0
    df.loc[ca_ngtv, 'ca'] = np.nan
    ca_more9 = df['ca'] > 9
    df.loc[ca_more9, 'ca'] = np.nan
    df['ca'].fillna(df['ca'].mean(), inplace=True)

#Thal
    thal_ngtv = df['thal'] < 0
    df.loc[thal_ngtv, 'thal'] = np.nan
    df['thal'].fillna(df['thal'].mean(), inplace=True)


#Num 
    df['num'].fillna(df['num'].max(), inplace=True)


#output
    print("\n")
    print("<------------- Data Info ------------->")
    df.info()

    return df
```

```
df = data_cleaning()
```
