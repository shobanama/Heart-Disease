```
heart_example = {"age":64,
                 "sex":0,
                 "cp":4,
                 "exang":1,
                 "oldpeak":1,
                 "slope":2,
                 "thal":7 }
```

```
from tensorflow.keras.models import load_model
import numpy as np
import joblib

heart_model = load_model('model.h5')
heart_scaler = joblib.load('scaler.pkl')

def return_prediction(model, scaler, sample_json):

      age = sample_json["age"]  
      sex = sample_json["sex"]
      cp = sample_json["cp"]
      exang = sample_json["exang"]
      oldpeak = sample_json["oldpeak"]
      slope = sample_json["slope"]
      thal = sample_json["thal"]

      heart = [[age,sex,cp,exang,oldpeak,slope,thal]]

      classes = np.array(['0', '1'])

      heart = scaler.transform(heart)

      class_ind = model.predict_classes(heart)[0]

      return classes[class_ind[0]]
```

```
return_prediction( heart_model, heart_scaler, heart_example)
```
