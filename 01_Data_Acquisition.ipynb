# Data Acquisition

```
import pandas as pd
from pandas import DataFrame
import numpy as np

from urllib.request import Request, urlopen
from urllib.error import  URLError
```

```
#Check Status/ Data Availability

req = Request('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/')
try:
    response = urlopen(req)
except URLError as e:
    if hasattr(e, 'reason'):
        print('We failed to reach a server.')
        print('Reason: ', e.reason)
    elif hasattr(e, 'code'):
        print('The server couldn\'t fulfill the request.')
        print('Error code: ', e.code)
else:
        print("Everything is fine")
```

```

clv = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data', names=["age",	"sex",	"cp",	"trestbps",	"chol",	"fbs",	"restecg",	"thalach",	"exang",	"oldpeak",	"slope",	'ca',	"thal",	"num"], na_values=['?'])
hug = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/reprocessed.hungarian.data',sep=' ', names=["age",	"sex",	"cp",	"trestbps",	"chol",	"fbs",	"restecg",	"thalach",	"exang",	"oldpeak",	"slope",	'ca',	"thal",	"num"], na_values=['?'])
swi = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.switzerland.data', names=["age",	"sex",	"cp",	"trestbps",	"chol",	"fbs",	"restecg",	"thalach",	"exang",	"oldpeak",	"slope",	'ca',	"thal",	"num"],  na_values=['?'])
va  = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.va.data', names=["age",	"sex",	"cp",	"trestbps",	"chol",	"fbs",	"restecg",	"thalach",	"exang",	"oldpeak",	"slope",	'ca',	"thal",	"num"],  na_values=['?'])

# df = clv.append(hug).append(swi).append(va) # old

df = pd.concat([clv,hug,swi,va]) # new
df.sample(20)
```
